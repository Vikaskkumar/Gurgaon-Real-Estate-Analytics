```python
df.insert(loc=3,column='sector',value=df['property_name'].str.split('in').str.get(1).str.replace('Gurgaon','').str.strip())
```

```python
df['sector'] = df['sector'].str.lower()
```

```python
df['sector'].value_counts()
```

```python
df[['sector']].to_csv('sector_data.csv', index=False)
```

```python
df.shape
```

```python

df['sector'] = df['sector'].str.replace('sohna road',           'sector 48',  regex=False)
df['sector'] = df['sector'].str.replace('nirvana country',      'sector 50',  regex=False)
df['sector'] = df['sector'].str.replace('palam vihar',          'sector 23',  regex=False)
df['sector'] = df['sector'].str.replace('dlf phase 2',          'sector 25',  regex=False)
df['sector'] = df['sector'].str.replace('dlf phase 1',          'sector 26',  regex=False)
df['sector'] = df['sector'].str.replace('sushant lok phase 1',  'sector 43',  regex=False)
df['sector'] = df['sector'].str.replace('laxman vihar phase 2', 'sector 4',   regex=False)
df['sector'] = df['sector'].str.replace('laxman vihar',         'sector 4',   regex=False)
df['sector'] = df['sector'].str.replace('dlf phase 4',          'sector 28',  regex=False)
df['sector'] = df['sector'].str.replace('dlf phase 3',          'sector 24',  regex=False)
df['sector'] = df['sector'].str.replace('sushant lok phase 3',  'sector 57',  regex=False)
df['sector'] = df['sector'].str.replace('dlf phase 5',          'sector 43',  regex=False)
df['sector'] = df['sector'].str.replace('gwal pahari',          'sector 53',  regex=False)
df['sector'] = df['sector'].str.replace('rajendra park',        'sector 105', regex=False)
df['sector'] = df['sector'].str.replace('uppals southend',      'sector 49',  regex=False)

mask = df['sector'].str.match(r'^sector[\s\-]?\d+', na=False)
df = df[mask]

```

```python
df.drop(columns=['property_name', 'address', 'description', 'rating'],inplace=True)
```
