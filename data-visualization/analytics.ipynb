```python
temp_df = df_merged.groupby('sector').agg(
    avg_price          = ('price', 'mean'),
    avg_price_per_sqft = ('price_per_sqft', 'mean'),
    avg_builtup_area   = ('built_up_area', 'mean'),
    latitude           = ('latitude', 'first'),
    longitude          = ('longitude', 'first')
).reset_index()

print(temp_df.head())
```

```python
import plotly.express as px

fig = px.scatter_map(
    temp_df.dropna(subset=['latitude', 'longitude']),
    lat='latitude',
    lon='longitude',
    hover_name='sector',
    hover_data={
        'avg_price':          True,
        'avg_price_per_sqft': True,
        'avg_builtup_area':   True,
        'latitude':           False,
        'longitude':          False
    },
    color='avg_price_per_sqft',
    size='avg_builtup_area',
    color_continuous_scale='Turbo',  # high contrast color scale
    size_max=30,                     # bigger bubbles
    zoom=11,
    height=600,
    title='Gurgaon Sectors Map'
)

fig.update_traces(marker=dict(opacity=0.85))  # slight transparency so overlaps show

fig.update_layout(map_style='open-street-map')
fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0})

fig.write_html("gurgaon_map.html")
print("Map saved! Open gurgaon_map.html in your browser.")
```

```python
import osmnx as ox
import geopandas as gpd
import plotly.express as px
import pandas as pd

# ── Step 1: Download Gurgaon sector boundaries from OpenStreetMap ──────────────
print("Downloading sector boundaries... (takes 1-2 min)")

sector_gdfs = []
for i in range(1, 116):
    try:
        gdf = ox.geocode_to_gdf(f"Sector {i}, Gurugram, Haryana, India")
        gdf['sector'] = i
        sector_gdfs.append(gdf)
        print(f"  ✓ Sector {i}")
    except:
        print(f"  ✗ Sector {i} not found")

# Combine all sectors into one GeoDataFrame
gurgaon_gdf = gpd.GeoDataFrame(pd.concat(sector_gdfs, ignore_index=True))

# ── Step 2: Merge with your temp_df ───────────────────────────────────────────
gurgaon_gdf['sector'] = gurgaon_gdf['sector'].astype(int)
temp_df['sector']     = temp_df['sector'].astype(int)

merged_gdf = gurgaon_gdf.merge(temp_df[['sector', 'avg_price', 'avg_price_per_sqft', 'avg_builtup_area']], 
                                on='sector', how='left')

# ── Step 3: Convert to GeoJSON for Plotly ─────────────────────────────────────
merged_gdf = merged_gdf.set_index('sector')
geojson    = merged_gdf.__geo_interface__

# ── Step 4: Plot ───────────────────────────────────────────────────────────────
fig = px.choropleth_map(
    merged_gdf.reset_index(),
    geojson=geojson,
    locations='sector',
    color='avg_price',
    hover_name='sector',
    hover_data={
        'avg_price':          True,
        'avg_price_per_sqft': True,
        'avg_builtup_area':   True,
    },
    color_continuous_scale='Turbo',   # green=cheap, red=expensive
    opacity=0.7,
    zoom=11,
    center={"lat": 28.4595, "lon": 77.0266},
    height=650,
    title='Gurgaon Sectors — Avg Price Map'
)

fig.update_layout(map_style='open-street-map')
fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0})

fig.write_html("gurgaon_sectors_price_map.html")
print("✓ Map saved! Open gurgaon_sectors_price_map.html in your browser.")
```
